```python
# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
x = np.outer(np.linspace(-3, 3, 32), np.ones(32))
y = x.copy().T # transpose
z = (np.sin(x **2) + np.cos(y **2) )
 
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
 
# Creating plot
ax.plot_surface(x, y, z)
 
# show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)

# fake up some data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
# Vertical Boxplot
green_diamond = dict(markerfacecolor='orange', marker='D')
fig3, ax3 = plt.subplots()
ax3.set_title('Vertical Boxplot')
ax3.boxplot(data, flierprops=green_diamond)
```

```python
# Horizontal Boxplot
green_diamond = dict(markerfacecolor='orange', marker='D')
fig3, ax3 = plt.subplots()
ax3.set_title('Horizontal Boxplot')
ax3.boxplot(data, vert = False,flierprops=green_diamond)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Y-axis values
y1 = [2, 3, 4.5]

# Y-axis values
y2 = [1, 1.5, 5]

# Function to add label and ticks
plt.tick_params (axis="x", labelsize=18, labelrotation=-60, labelcolor="turquoise") 
plt.tick_params (axis="y", labelsize=12, labelrotation=20, labelcolor="orange")



# Function add a Grid
plt.grid(True)

# Function to plot
plt.plot(y1)
plt.plot(y2)

# Function add a legend
plt.legend(["blue", "green"], loc ="lower right")

# function to show the plot
plt.show()
```

```python
# importing mplot3d toolkits, numpy and matplotlib
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

# syntax for 3-D projection
ax = plt.axes(projection ='3d')

# defining all 3 axes
z = np.linspace(0, 1, 1000)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')
plt.show()
```
